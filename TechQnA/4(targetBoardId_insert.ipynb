```python
import json

# JSON 파일 읽기
with open('board_posts_imgConvert_attachConvert.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# targetBoardId 값 설정
target_board_id = "1433257924444065792" # Tech-Q&A

# 모든 게시글의 targetBoardId 업데이트
for post in data['posts']:
    post['targetBoardId'] = target_board_id

# 업데이트된 내용을 새 파일로 저장
with open('board_targetBoardId_insert.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)

print(f"완료! {len(data['posts'])}개의 게시글 업데이트됨")
```

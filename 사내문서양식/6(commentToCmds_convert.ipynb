```python
import json

# JSON 파일 읽기
with open('board_comments.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 결과를 저장할 파일 열기
with open('converted_comments.txt', 'w', encoding='utf-8') as output:
    # 각 comment를 addPostComment{...} 형식으로 변환
    for comment in data['comments']:
        # JSON을 한 줄로 직렬화 (공백 없이)
        comment_json = json.dumps(comment, ensure_ascii=False, separators=(',', ':'))
        # addPostComment 접두사와 함께 출력
        output.write(f'addPostComment{comment_json}\n')

print(f"변환 완료! 총 {len(data['comments'])}개의 댓글을 변환했습니다.")
print("결과 파일: converted_comments.txt")
```

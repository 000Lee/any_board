```python
# 재크롤링시 문서 중복을 막기 위한 폴더 삭제

import shutil
import os

# 크롤링 시작 전에 실행
folders_to_clean = ['DB업로드로 변경 test _cp표준단가']

for folder in folders_to_clean:
    folder_path = folder 
    
    if os.path.exists(folder_path):
        shutil.rmtree(folder_path)  # 폴더와 모든 내용물 삭제
        print(f'{folder} 폴더 삭제 완료')
    
```

```python
#ì²¨ë¶€íŒŒì¼ ì²˜ë¦¬ - íŒŒì¼ ë³µì‚¬ ë° json path ë³€ê²½

import json
import os
import shutil
from pathlib import Path

# ==================== ì„¤ì • ì˜ì—­ (ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”) ====================

# ì²¨ë¶€íŒŒì¼ ì›ë³¸ ê²½ë¡œë“¤ - ì—¬ëŸ¬ ê²½ë¡œ ì¶”ê°€ ê°€ëŠ¥
BASE_FILE_PATHS = [
    r"C:\Users\LEEJUHWAN\Downloads\2021-01-01~2025-10-31\html\resource\file\brd",
    r"C:\Users\LEEJUHWAN\Downloads\2016-01-01~2020-12-31\html\resource\file\brd",
    r"C:\Users\LEEJUHWAN\Downloads\2011-01-01~2015-12-31\html\resource\file\brd",
    r"C:\Users\LEEJUHWAN\Downloads\2010-01-01~2010-12-31\html\resource\file\brd",
    # ë” ì¶”ê°€í•˜ê³  ì‹¶ìœ¼ë©´ ì—¬ê¸°ì— ê³„ì† ì¶”ê°€

]

# ì²¨ë¶€íŒŒì¼ ë³µì‚¬ë³¸ ì €ì¥ í´ë”
OUTPUT_FILE_FOLDER = r"C:\Users\LEEJUHWAN\Downloads\board_hredu_attachments"

# ì²˜ë¦¬í•  JSON íŒŒì¼
JSON_FILE = r"board_posts_imgConvert.json"

# pathì—ì„œ ì œê±°í•  ì ‘ë‘ì‚¬ (ì´ ë¶€ë¶„ì„ ì œê±°í•˜ê³  ìƒëŒ€ê²½ë¡œë¡œ ë³€ê²½)
REMOVE_PREFIX = r"C:\Users\LEEJUHWAN\Downloads"

# ======================================================================


def find_attachment_folder(board_id):
    """ì—¬ëŸ¬ ê²½ë¡œì—ì„œ ê²Œì‹œíŒ IDì— í•´ë‹¹í•˜ëŠ” ì²¨ë¶€íŒŒì¼ í´ë” ì°¾ê¸°"""
    for base_path in BASE_FILE_PATHS:
        attachment_folder = Path(base_path) / board_id
        if attachment_folder.exists():
            return attachment_folder
    return None


def find_file_in_folder(folder, filename):
    """í´ë” ë‚´ì—ì„œ íŒŒì¼ëª…ê³¼ ì¼ì¹˜í•˜ëŠ” íŒŒì¼ ì°¾ê¸° (ëŒ€ì†Œë¬¸ì êµ¬ë¶„ ì—†ìŒ)"""
    if not folder or not folder.exists():
        return None
    
    # ì •í™•í•œ íŒŒì¼ëª…ìœ¼ë¡œ ë¨¼ì € ì°¾ê¸°
    target_file = folder / filename
    if target_file.exists() and target_file.is_file():
        return target_file
    
    # ëŒ€ì†Œë¬¸ì êµ¬ë¶„ ì—†ì´ ì°¾ê¸°
    for file in folder.iterdir():
        if file.is_file() and file.name.lower() == filename.lower():
            return file
    
    return None


def process_single_post_attachments(data, post_index):
    """í•˜ë‚˜ì˜ postì˜ ì²¨ë¶€íŒŒì¼ ì²˜ë¦¬"""
    
    # sourceIdì—ì„œ ê²Œì‹œíŒID ì¶”ì¶œ
    source_id = data.get('sourceId', '')
    if not source_id:
        print(f"âš ï¸ [{post_index}] sourceId ì—†ìŒ, ê±´ë„ˆëœ€")
        return data
    
    board_id = source_id.lstrip('_')
    
    # attaches ë°°ì—´ í™•ì¸
    attaches = data.get('attaches', [])
    if not attaches:
        return data  # ì²¨ë¶€íŒŒì¼ ì—†ìœ¼ë©´ ê·¸ëŒ€ë¡œ ë°˜í™˜
    
    # ì—¬ëŸ¬ ê²½ë¡œì—ì„œ ì²¨ë¶€íŒŒì¼ í´ë” ì°¾ê¸°
    attachment_folder = find_attachment_folder(board_id)
    
    if not attachment_folder:
        print(f"âš ï¸ [{post_index}] ê²Œì‹œíŒID {board_id} í´ë” ì—†ìŒ (ëª¨ë“  ê²½ë¡œ ê²€ìƒ‰ ì™„ë£Œ)")
        # í´ë”ê°€ ì—†ìœ¼ë©´ ëª¨ë“  ì²¨ë¶€íŒŒì¼ì˜ pathë¥¼ ê³µë€ìœ¼ë¡œ
        for attach in attaches:
            attach['path'] = ''
        return data
    
    # ë³µì‚¬ë³¸ ì €ì¥í•  ê²Œì‹œíŒë³„ í´ë” ìƒì„±
    output_board_folder = Path(OUTPUT_FILE_FOLDER) / f"brd{board_id}"
    output_board_folder.mkdir(parents=True, exist_ok=True)
    
    print(f"\nğŸ“„ [{post_index}] ê²Œì‹œíŒID: {board_id}")
    print(f"   ì›ë³¸: {attachment_folder}")
    print(f"   ë³µì‚¬: {output_board_folder}")
    print(f"   âœ… ì²¨ë¶€íŒŒì¼ {len(attaches)}ê°œ ë°œê²¬")
    
    # ê° ì²¨ë¶€íŒŒì¼ ì²˜ë¦¬
    processed_attaches = []
    copied_count = 0
    
    for idx, attach in enumerate(attaches):
        filename = attach.get('name', '')
        if not filename:
            print(f"      âš ï¸ [{idx}] íŒŒì¼ëª… ì—†ìŒ - path ê³µë€ ì²˜ë¦¬")
            attach['path'] = ''
            processed_attaches.append(attach)
            continue
        
        # ì›ë³¸ í´ë”ì—ì„œ íŒŒì¼ ì°¾ê¸°
        source_file = find_file_in_folder(attachment_folder, filename)
        
        if source_file:
            # ë³µì‚¬ë³¸ íŒŒì¼ ê²½ë¡œ
            dest_file = output_board_folder / filename
            
            # íŒŒì¼ ë³µì‚¬
            try:
                shutil.copy2(source_file, dest_file)
                copied_count += 1
                print(f"      âœ… {filename}")
                
                # pathë¥¼ ìƒëŒ€ê²½ë¡œë¡œ ë³€ê²½ (REMOVE_PREFIX ì œê±°)
                new_path = str(dest_file).replace('\\', '/')
                if REMOVE_PREFIX:
                    new_path = new_path.replace(REMOVE_PREFIX.replace('\\', '/'), '').lstrip('/')
                    new_path = new_path.replace(REMOVE_PREFIX, '').lstrip('\\')
                
                attach['path'] = new_path
                
            except Exception as e:
                print(f"      âŒ {filename} ë³µì‚¬ ì‹¤íŒ¨: {e}")
                attach['path'] = ''
        else:
            print(f"      âš ï¸ íŒŒì¼ ì—†ìŒ: {filename} - path ê³µë€ ì²˜ë¦¬")
            attach['path'] = ''
        
        processed_attaches.append(attach)
    
    print(f"   ğŸ“Š ë³µì‚¬ ì™„ë£Œ: {copied_count}/{len(attaches)}ê°œ")
    
    # ìˆ˜ì •ëœ attaches ì ìš©
    data['attaches'] = processed_attaches
    return data


def process_json_file(json_file_path):
    """JSON íŒŒì¼ ì²˜ë¦¬"""
    
    # ì¶œë ¥ í´ë” ìƒì„±
    os.makedirs(OUTPUT_FILE_FOLDER, exist_ok=True)
    
    # 1. JSON íŒŒì¼ ì½ê¸°
    print(f"ğŸ“‚ íŒŒì¼ ì½ëŠ” ì¤‘: {json_file_path}")
    print(f"ğŸ“‚ ê²€ìƒ‰í•  ì²¨ë¶€íŒŒì¼ ê²½ë¡œ: {len(BASE_FILE_PATHS)}ê°œ\n")
    
    with open(json_file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    
    posts = data.get('posts', [])
    total_count = len(posts)
    
    print(f"âœ… ì´ {total_count}ê°œì˜ posts ë°œê²¬\n")
    
    # 2. ê° post ì²˜ë¦¬
    processed_posts = []
    success_count = 0
    total_attachments = 0
    
    for idx, post in enumerate(posts, 1):
        try:
            # ì²¨ë¶€íŒŒì¼ì´ ìˆëŠ” ê²½ìš°ë§Œ ì¹´ìš´íŠ¸
            if post.get('attaches'):
                total_attachments += len(post.get('attaches'))
            
            result = process_single_post_attachments(post, idx)
            processed_posts.append(result)
            
            if result != post:  # ë³€ê²½ì‚¬í•­ì´ ìˆìœ¼ë©´
                success_count += 1
        except Exception as e:
            print(f"âŒ [Post {idx}] ì²˜ë¦¬ ì˜¤ë¥˜: {e}")
            processed_posts.append(post)  # ì˜¤ë¥˜ ì‹œ ì›ë³¸ ìœ ì§€
            continue
    
    # 3. ìˆ˜ì •ëœ ë°ì´í„° ì €ì¥
    output_file = json_file_path.replace('.json', '_attachConvert.json')
    
    output_data = {"posts": processed_posts}
    
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(output_data, f, ensure_ascii=False, indent=2)
    
    print(f"\n{'='*60}")
    print(f"âœ… ì²˜ë¦¬ ì™„ë£Œ!")
    print(f"   - ì²¨ë¶€íŒŒì¼ ìˆëŠ” posts: {success_count}ê°œ")
    print(f"   - ì´ ì²¨ë¶€íŒŒì¼ ê°œìˆ˜: {total_attachments}ê°œ")
    print(f"   - ë³µì‚¬ëœ íŒŒì¼ ìœ„ì¹˜: {OUTPUT_FILE_FOLDER}")
    print(f"   - ìˆ˜ì •ëœ JSON íŒŒì¼: {output_file}")
    print(f"{'='*60}")


# ì‹¤í–‰
if __name__ == "__main__":
    process_json_file(JSON_FILE)
```

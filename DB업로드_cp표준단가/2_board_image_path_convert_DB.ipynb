```python
# ì´ë¯¸ì§€ ê²½ë¡œ ë³€í™˜ - DB ì§ì ‘ ì—°ë™ ë²„ì „
# type ì»¬ëŸ¼ìœ¼ë¡œ ê²Œì‹œíŒ ì„ íƒ

import json
import os
import shutil
import re
from pathlib import Path
import pymysql
from sqlalchemy import create_engine

# ==================== DB ì„¤ì • ====================
USER = "root"
PASS = "1234"
HOST = "127.0.0.1"
PORT = 3306
DB = "crawling"

# ==================== ì„¤ì • ì˜ì—­ (ì—¬ê¸°ë¥¼ ìˆ˜ì •í•˜ì„¸ìš”) ====================

# ê²Œì‹œíŒ íƒ€ì… (í•„í„°ë§ìš©)
BOARD_TYPE = "cp"  # âœ¨ ì¶”ê°€: hredu, cp, sales ë“±

# ì´ë¯¸ì§€ ì›ë³¸ ê²½ë¡œë“¤ - ì—¬ëŸ¬ ê²½ë¡œ ì¶”ê°€ ê°€ëŠ¥
BASE_IMAGE_PATHS = [
    r"C:\Users\LEEJUHWAN\Downloads\2021-01-01~2025-10-31\html\resource\image\namoeditor",
    r"C:\Users\LEEJUHWAN\Downloads\2021-01-01~2025-10-31\html\resource\image\daumeditor",
    r"C:\Users\LEEJUHWAN\Downloads\2016-01-01~2020-12-31\html\resource\image\namoeditor",
    r"C:\Users\LEEJUHWAN\Downloads\2016-01-01~2020-12-31\html\resource\image\daumeditor",
    # ë” ì¶”ê°€í•˜ê³  ì‹¶ìœ¼ë©´ ì—¬ê¸°ì— ê³„ì† ì¶”ê°€
]

# ì´ë¯¸ì§€ ë³µì‚¬ë³¸ ì €ì¥ í´ë” (íƒ€ì…ë³„ë¡œ ìë™ ìƒì„±)
OUTPUT_IMAGE_FOLDER = rf"C:\Users\LEEJUHWAN\Downloads\board_{BOARD_TYPE}_img"  # âœ¨ íƒ€ì… ë°˜ì˜

# srcì—ì„œ ì œê±°í•  ì ‘ë‘ì‚¬ (ì´ ë¶€ë¶„ì„ ì œê±°í•˜ê³  ìƒëŒ€ê²½ë¡œë¡œ ë³€ê²½)
REMOVE_PREFIX = r"C:\Users\LEEJUHWAN\Downloads"

# ======================================================================


def find_image_folder(folder_name):
    """ì—¬ëŸ¬ ê²½ë¡œì—ì„œ ì´ë¯¸ì§€ í´ë” ì°¾ê¸°"""
    for base_path in BASE_IMAGE_PATHS:
        image_folder = Path(base_path) / folder_name
        if image_folder.exists():
            return image_folder
    return None


def process_single_post(data, post_index):
    """í•˜ë‚˜ì˜ post ë°ì´í„° ì²˜ë¦¬"""
    
    # sourceIdì—ì„œ ë¬¸ì„œID ì¶”ì¶œ
    source_id = data.get('sourceId', '')
    if not source_id:
        print(f"âš ï¸ [{post_index}] sourceId ì—†ìŒ, ê±´ë„ˆëœ€")
        return data  # ì›ë³¸ ê·¸ëŒ€ë¡œ ë°˜í™˜
    
    doc_id = source_id.lstrip('_')
    folder_name = f"brd{doc_id}"
    
    # ì—¬ëŸ¬ ê²½ë¡œì—ì„œ ì´ë¯¸ì§€ í´ë” ì°¾ê¸°
    image_folder = find_image_folder(folder_name)
    
    if not image_folder:
        print(f"âš ï¸ [{post_index}] í´ë” ì—†ìŒ: {folder_name} (ëª¨ë“  ê²½ë¡œ ê²€ìƒ‰ ì™„ë£Œ)")
        return data  # ì›ë³¸ ê·¸ëŒ€ë¡œ ë°˜í™˜
    
    # ë³µì‚¬ë³¸ ì €ì¥í•  ë¬¸ì„œë³„ í´ë” ìƒì„±
    output_doc_folder = Path(OUTPUT_IMAGE_FOLDER) / folder_name
    output_doc_folder.mkdir(parents=True, exist_ok=True)
    
    print(f"\nğŸ“„ [{post_index}] ë¬¸ì„œID: {doc_id}")
    print(f"   ì›ë³¸: {image_folder}")
    print(f"   ë³µì‚¬: {output_doc_folder}")
    
    # HTMLì—ì„œ img íƒœê·¸ ì°¾ê¸°
    html_content = data.get('content', '')
    if not html_content:
        print(f"   âš ï¸ content ì—†ìŒ")
        return data
    
    img_pattern = r'<img[^>]+src="([^"]+)"[^>]*>'
    img_matches = list(re.finditer(img_pattern, html_content))
    
    if not img_matches:
        print(f"   â„¹ï¸ ì´ë¯¸ì§€ ì—†ìŒ")
        return data
    
    print(f"   âœ… ì´ë¯¸ì§€ {len(img_matches)}ê°œ ë°œê²¬")
    
    # íŒŒì¼ ë³µì‚¬ & src ê²½ë¡œ ë³€ê²½
    new_html = html_content
    copied_count = 0
    
    for idx, match in enumerate(img_matches):
        old_src = match.group(1)
        
        # ì›ë³¸ íŒŒì¼ (0, 1, 2...)
        old_file = image_folder / str(idx)
        
        # ë³µì‚¬ë³¸ íŒŒì¼ (.jpg í™•ì¥ì ì¶”ê°€)
        new_file = output_doc_folder / f"{idx}.jpg"
        
        # íŒŒì¼ ë³µì‚¬
        if old_file.exists():
            shutil.copy2(old_file, new_file)
            copied_count += 1
            print(f"      âœ… {old_file.name} â†’ {new_file.name}")
            
            # src ê²½ë¡œ ë³€ê²½ (ìƒëŒ€ê²½ë¡œë¡œ)
            new_src = str(new_file).replace('\\', '/')
            if REMOVE_PREFIX:
                prefix_with_slash = REMOVE_PREFIX.replace('\\', '/')
                if new_src.startswith(prefix_with_slash):
                    new_src = new_src[len(prefix_with_slash):].lstrip('/')
            
            new_html = new_html.replace(old_src, new_src, 1)
        else:
            print(f"      âš ï¸ íŒŒì¼ ì—†ìŒ: {old_file.name} - src ê³µë€ ì²˜ë¦¬")
            # srcë¥¼ ë¹ˆ ë¬¸ìì—´ë¡œ ë³€ê²½
            new_html = new_html.replace(old_src, '', 1)
    
    print(f"   ğŸ“Š ë³µì‚¬ ì™„ë£Œ: {copied_count}/{len(img_matches)}ê°œ")
    
    # ìˆ˜ì •ëœ HTML ì ìš©
    data['content'] = new_html
    return data


def process_db(board_type):
    """DBì—ì„œ íŠ¹ì • typeì˜ ë°ì´í„°ë§Œ ì½ê³  ì²˜ë¦¬ í›„ DBì— ì—…ë°ì´íŠ¸"""
    
    # ì¶œë ¥ í´ë” ìƒì„±
    os.makedirs(OUTPUT_IMAGE_FOLDER, exist_ok=True)
    
    # DB ì—°ê²°
    print(f"ğŸ“‚ DB ì—°ê²° ì¤‘...")
    print(f"ğŸ“‚ ê²Œì‹œíŒ íƒ€ì…: {board_type}")  # âœ¨ ì¶”ê°€
    print(f"ğŸ“‚ ê²€ìƒ‰í•  ì´ë¯¸ì§€ ê²½ë¡œ: {len(BASE_IMAGE_PATHS)}ê°œ\n")
    
    connection = pymysql.connect(
        host=HOST,
        user=USER,
        password=PASS,
        database=DB,
        port=PORT,
        charset='utf8mb4'
    )
    
    try:
        with connection.cursor(pymysql.cursors.DictCursor) as cursor:
            # 1. DBì—ì„œ íŠ¹ì • typeì˜ ë°ì´í„°ë§Œ ì½ê¸°
            query = "SELECT * FROM board_posts WHERE type = %s"  # âœ¨ WHERE ì¡°ê±´ ì¶”ê°€
            cursor.execute(query, (board_type,))
            posts = cursor.fetchall()
            
            total_count = len(posts)
            print(f"âœ… ì´ {total_count}ê°œì˜ posts ë°œê²¬\n")
            
            # 2. ê° post ì²˜ë¦¬
            processed_count = 0
            
            for idx, post in enumerate(posts, 1):
                try:
                    # ì›ë³¸ content ì €ì¥ (ë¹„êµìš©)
                    original_content = post.get('content', '')
                    
                    # attaches_jsonì„ íŒŒì‹±í•˜ì—¬ attachesë¡œ ë³€í™˜
                    if post.get('attaches_json'):
                        post['attaches'] = json.loads(post['attaches_json'])
                    else:
                        post['attaches'] = []
                    
                    # ê¸°ì¡´ ë¡œì§ ê·¸ëŒ€ë¡œ ì‚¬ìš©
                    result = process_single_post(post, idx)
                    
                    # contentê°€ ë³€ê²½ëœ ê²½ìš°ë§Œ DB ì—…ë°ì´íŠ¸
                    if result['content'] != original_content:
                        update_query = """
                            UPDATE board_posts 
                            SET content = %s 
                            WHERE sourceId = %s AND type = %s
                        """  # âœ¨ WHERE ì¡°ê±´ì— typeë„ ì¶”ê°€
                        cursor.execute(update_query, (result['content'], result['sourceId'], board_type))
                        processed_count += 1
                    
                except Exception as e:
                    print(f"âŒ [Post {idx}] ì²˜ë¦¬ ì˜¤ë¥˜: {e}")
                    continue
            
            # ë³€ê²½ì‚¬í•­ ì»¤ë°‹
            connection.commit()
            
            print(f"\n{'='*60}")
            print(f"âœ… ì²˜ë¦¬ ì™„ë£Œ!")
            print(f"   - ê²Œì‹œíŒ íƒ€ì…: {board_type}")  # âœ¨ ì¶”ê°€
            print(f"   - ì²˜ë¦¬ëœ posts: {processed_count}/{total_count}ê°œ")
            print(f"   - ë³µì‚¬ëœ ì´ë¯¸ì§€ ìœ„ì¹˜: {OUTPUT_IMAGE_FOLDER}")
            print(f"   - DB í…Œì´ë¸”: board_posts (content ì»¬ëŸ¼ ì—…ë°ì´íŠ¸)")
            print(f"{'='*60}")
    
    finally:
        connection.close()
        print("\nâœ… DB ì—°ê²° ì¢…ë£Œ")


# ì‹¤í–‰
if __name__ == "__main__":
    process_db(BOARD_TYPE)  # âœ¨ board_type ì¸ì ì¶”ê°€
```
